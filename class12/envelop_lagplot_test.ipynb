```python
from scipy.signal import hilbert
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(44100)/44100
s1 = np.sin(2*np.pi*5*t)
s2 = np.sin(2*np.pi*10*t)
# s = np.hstack((s1,s2))
s = s1*s2
```

```python
plt.plot(s)
```

```python
# s = np.random.rand(44100)
v = np.abs( hilbert(s) )
v = np.abs( s )
# plt.plot(v[:100] - s[:100])
plt.plot(s)
plt.plot(v)
```

```python
# lag plots
for lag in range(500, 3000, 100):
  p1 = s[:-lag]
  p2 = s[lag:]
  p3 = s[:-lag//2]
  p4 = s[lag//2:]
  plt.clf()
  plt.plot( p1, p2 )
  plt.plot( p3, p4 )
  plt.pause(0.01)
```
